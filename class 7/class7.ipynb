```python
import json
import re
```

```python
corpus = json.load(open('lyrics.json'))
```

```python
#1. text we want to find
query = "knees heavy palms sweaty aweater already moms spaguetti"
```

```python
# 2. tokenise the query
query_tokens = re.findall(r"\w\w\w+", query)
```

```python
# 3. build an inverted index { word: [song_title] }
from collections import defaultdict
index = defaultdict(set) # dict but with empty lists as default values
for doc in corpus:
    for word in re.findall(r"\w\w\w+", doc["text"].lower()):
        index[word].add(doc["title"])
```

```python
# 4. find song tittles matching the query
from collections import Counter
query = "shot sheriff not deputy"
query_tokens = re.findall(r"\w\w\w+", query)

result = []
for word in query_tokens:
    titles = index.get(word, []) # return empty list if word not in index
    print('titles', titles)
    result += titles
    print('result', result)


Counter(result).most_common(10)
```
